```python
import numpy as np
from matplotlib import pyplot as plt
from numpy import genfromtxt
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
%matplotlib inline
```

## Load data

```python
# Upload problem3data.txt to a 'data' folder in Google Colab, using the "Files"
# tab on the left.
# (You can also use urllib.request.urlretrieve, as in set 1)
import urllib.request
urllib.request.urlretrieve('https://raw.githubusercontent.com/charlesincharge/Caltech-CS155-2022/main/sets/set2/data/problem3data.txt', 'problem3data.txt')

train_file = 'problem3data.txt'
train_data = genfromtxt(train_file, delimiter='\t')

y_train = train_data[:, 9]
X_train = train_data[:, :9]
```

```python
lasso_alphas = np.linspace(0.01, 3, 30)
lasso_weights = []

for a in lasso_alphas:
  # Instantiate a Lasso regression object:
  lasso = Lasso(alpha = a)
  lasso.fit(X_train, y_train)   # Fit the lasso regression model to the data
  lasso_weights.append(lasso.coef_)
```

```python
plt.figure()

for weight in np.transpose(lasso_weights):
  plt.plot(lasso_alphas, weight, marker = 'o', linewidth = 0)

plt.legend([str(a+1)+" weight" for a in range(len(np.transpose(lasso_weights)))], loc = 'upper right')
plt.xlabel('Alphas')
plt.ylabel('Weight Values')
plt.title('Lasso Regression')

plt.show()
```

```python
ridge_alphas = range(1,10001)
ridge_weights = []

for a in ridge_alphas:
  # Instantiate a Ridge regression object:
  ridge = Ridge(alpha = a)
  ridge.fit(X_train, y_train)   # Fit the ridge regression model to the data
  ridge_weights.append(ridge.coef_)
```

```python
plt.figure()

for weight in np.transpose(ridge_weights):
  plt.plot(ridge_alphas, weight, marker = 'o', linewidth = 0)

plt.legend([str(a+1)+" weight" for a in range(len(np.transpose(ridge_weights)))], loc = 'upper right')
plt.xlabel('Alphas')
plt.ylabel('Weight Values')
plt.title('Ridge Regression')

plt.show()
```
